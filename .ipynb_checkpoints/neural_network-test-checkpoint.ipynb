```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
import os
import sys
from glob import glob
from sklearn.neural_network import MLPRegressor
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

home_directory = os.getenv('HOME')
dStar_directory = home_directory + '/dStar/'
sys.path.append(dStar_directory + 'tools/')

import reader
```

```python
def read_parameters(inlist_file, parameter_names):
    
    inlist = open(inlist_file, 'r')
    
    parameter_values = np.zeros(len(parameter_names))
    
    for line in inlist.readlines():
        
        for i, parameter_name in enumerate(parameter_names):
                if line[:7+len(parameter_name)] == '    {} = '.format(parameter_name):
                    parameter_values[i] = float(line[7+len(parameter_name):])
                    break
                    
    return parameter_values
```

```python
def lightcurve_interpolation(data_file, time_range):
    f = reader.dStarReader(data_file)
    
    dStar_times = f.data.time
    dStar_luminosities = f.data.lg_Lsurf
    
    interpolation = interpolate.interp1d(dStar_times, dStar_luminosities)
    
    return interpolation(time_range)
```

```python
def create_data_set(LOGS_directories, time_range):
    
    luminosities = np.zeros((len(LOGS_directories), len(time_range)))
    
    for i, LOGS_directory in enumerate(LOGS_directories):
    
        data_file = LOGS_directory + 'history.data'

        luminosities[i] = lightcurve_interpolation(data_file, time_range)
        
    return luminosities
```

```python
# Set directory variables to load data.
training_LOGS_directories = sorted(glob('/home/justin/Storage/Research/neutron_star_code/basic_run/training_LOGS/LOGS????/'))
testing_LOGS_directories = sorted(glob('/home/justin/Storage/Research/neutron_star_code/basic_run/testing_LOGS/LOGS????/'))

training_parameters = np.load('training_parameters.npy')
testing_parameters = np.load('testing_parameters.npy')
parameter_names = np.load('parameter_names.npy')

# Set the base time range to the time range on the first training model.
# We will interpolate the luminosities of all other models to these values of time.
f = reader.dStarReader(training_LOGS_directories[0] + 'history.data')
time_range = f.data.time

training_luminosities = create_data_set(training_LOGS_directories, time_range)
testing_luminosities = create_data_set(testing_LOGS_directories, time_range)
```

```python
scaler = StandardScaler()
scaler.fit(training_parameters)
training_parameters_scaled = scaler.transform(training_parameters)
testing_parameters_scaled = scaler.transform(testing_parameters)

pca = PCA(n_components = 2)

training_luminosities_pca = pca.fit_transform(training_luminosities)
testing_luminosities_pca = pca.transform(testing_luminosities)
```

```python
X_train = np.copy(training_parameters_scaled)
y_train = np.copy(training_luminosities_pca)
X_test = np.copy(testing_parameters_scaled)
y_test = np.copy(testing_luminosities_pca)

regr = MLPRegressor(hidden_layer_sizes=(30, 20, 10), max_iter=500, learning_rate='adaptive').fit(X_train, y_train)
y_pred = regr.predict(X_test)

predicted_luminosities = pca.inverse_transform(y_pred)
```

```python
L2_errors = np.sqrt(np.sum(((testing_luminosities - predicted_luminosities)/testing_luminosities)**2, axis=1)/testing_luminosities.shape[1])

print(np.mean(L2_errors))
print(np.max(L2_errors))
print(np.std(L2_errors))
```

```python
fig = plt.figure()
fig.patch.set_facecolor('white')
fig.set_dpi(200)

for i, test_luminosities in enumerate(testing_luminosities[:5]):

    p = plt.plot(time_range, test_luminosities, label='Mass = {0:.2f}'.format(testing_parameters[i,0]) + ', Radius = {0:.2f}'.format(testing_parameters[i,1]))
    color = p[0].get_color()
    plt.plot(time_range, predicted_luminosities[i], '--', color=color)
    
    
plt.xlabel('Time')
plt.ylabel('Log (L)')
plt.legend()
plt.show()
```

```python
fig = plt.figure()
fig.patch.set_facecolor('white')
fig.set_dpi(200)

for i, test_luminosities in enumerate(testing_luminosities[:5]):

    plt.plot(time_range, (predicted_luminosities[i] - test_luminosities)/test_luminosities)
    
    
plt.xlabel('Time')
plt.ylabel('Relative Error in Log (L)')
plt.show()
```
