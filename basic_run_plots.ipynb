```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
from glob import glob

home_directory = os.getenv('HOME')
dStar_directory = home_directory + '/dStar/'
sys.path.append(dStar_directory + 'tools/')

import reader
```

```python
# Set directory variables to load data.

# The LOGS directory containing the profiles of the run of interest. In this case, we are looking at the basic run.
LOGS_directories = sorted(glob('LOGS????/'))

fig = plt.figure()
fig.patch.set_facecolor('white')
fig.set_dpi(200)

for LOGS_directory in LOGS_directories:

    f = reader.dStarReader(LOGS_directory + 'history.data')
    
    core_mass = f.header.core_mass
    t = f.data.time
    lgL = f.data.lg_Lsurf
    
    '''fig = plt.figure()
    fig.patch.set_facecolor('white')
    fig.set_dpi(200)'''

    plt.semilogy(t, lgL)
    
plt.xlabel('Time')
plt.ylabel('Log (L)')
plt.title(r'Core Mass: {0:.2f} M$_\odot$'.format(core_mass))
plt.show()
```
